## coloring the edges

```python
nodes = list(range(10))
```

```python
edges = [(0, 1, {"color": "r"}),
        (0, 2, {"color": "b"}),
        (0, 3, {"color": "b"}),
        (0, 4, {"color": "b"}),
        (0, 5, {"color": "b"}),
        (0, 6, {"color": "b"}),
        (0, 7, {"color": "b"}),
        (0, 8, {"color": "b"}),
        (0, 9, {"color": "b"}),]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
labels = dict((n, n) for n in nodes)
```

```python
## set position as spring
pos = nx.spring_layout(G)
```

```python
# nodes
options = {"node_size": 400, "alpha": 0.5}
nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_color="b", **options)

# edges
for a, b in G.edges:
    nx.draw_networkx_edges(G, pos, edgelist=[(a, b)], width=1.0, alpha=0.5, edge_color=[G[a][b]["color"]])

# labels
nx.draw_networkx_labels(G, pos, labels, font_size=16)

plt.axis("off")
plt.show()
```

## generated numbers

using spiral layout

```python
import numpy as np
import time
```

```python
n = 5000
```

```python
nodes = list(range(n))
```

```python
labels = dict((node, node) for node in nodes)
```

```python
edges = [(i, j) for i in range(n) for j in np.random.randint(low=0, high=100, size = 10)]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
pos = nx.spiral_layout(G)
```

```python
start = time.time()
plt.figure(figsize=(10,10))
nx.draw(G, pos, alpha=0.3, with_labels=True)

print(time.time() - start, "seconds")
```

using spring layout

```python
pos = nx.spring_layout(G)
```

```python
start = time.time()
plt.figure(figsize=(12,12))
nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_size=100)
nx.draw_networkx_edges(G, pos, edgelist=edges, width=1, alpha=0.5)
nx.draw_networkx_labels(G, pos, labels, font_size=6)
plt.axis("off")
plt.show()
```

```python
print(time.time() - start , "seconds")
```

using bipartite layout

```python
G = nx.bipartite.gnmk_random_graph(3, 5, 10, seed=123)
```

```python
G.nodes
```

```python
G.edges
```

```python
top = nx.bipartite.sets(G)[0]
```

```python
pos = nx.bipartite_layout(G, top)
```

```python
start = time.time()
plt.figure(figsize=(10,10))
nx.draw(G, pos, alpha=0.3, with_labels=True)

print(time.time() - start, "seconds")
```

using circular layout

```python
pos = nx.circular_layout(G)
```

```python
start = time.time()
plt.figure(figsize=(10,10))
nx.draw(G, pos, alpha=0.3, with_labels=True)
```

```python
print(time.time() - start, "seconds")
```
